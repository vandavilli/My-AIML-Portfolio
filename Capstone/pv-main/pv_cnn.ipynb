```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from gensim.models import KeyedVectors
from keras.preprocessing.text import Tokenizer
from keras.models import Sequential
from keras.layers import Dense, Embedding, Conv1D, MaxPooling1D, GlobalMaxPooling1D
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, KFold
```

```python
train_data = pd.read_csv('data/train.csv')
test_data = pd.read_csv('data/test.csv')
dev_data = pd.read_csv('data/dev.csv')

dataframes = [train_data, test_data]
df = process_load_data(dataframes)
```

```python
model_path = './biowordvec/BioWordVec_PubMed_MIMICIII_d200.vec.bin'
bio_word_vec_model = load_wordvec_model(model_path)
```

```python
def word_2_vec(sentence):
    words = sentence.split()
    word_vecs = [bio_word_vec_model[word] for word in words if word in bio_word_vec_model]
    return np.mean(word_vecs, axis=0) if len(word_vecs) else np.zeros(bio_word_vec_model.vector_size)
```

```python
df['vec'] = df['context'].apply(word_2_vec)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('sentiment', axis=1), 
                                                    df['sentiment'], test_size=0.2, random_state=42)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))

X_train_embeddings = np.array(X_train['vec'].tolist())
X_test_embeddings = np.array(X_test['vec'].tolist())

X_train_scaled = scaler.fit_transform(X_train_embeddings)
X_test_scaled = scaler.transform(X_test_embeddings)
```

```python
def create_model(input_shape):
    model = Sequential()
    model.add(Conv1D(128, 5, activation='relu', input_shape=input_shape))
    model.add(MaxPooling1D(5))
    model.add(Conv1D(128, 5, activation='relu'))
    model.add(GlobalMaxPooling1D())
    model.add(Dense(128, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
X_train_scaled = X_train_scaled.reshape(X_train_scaled.shape[0], X_train_scaled.shape[1], 1)
X_test_scaled = X_test_scaled.reshape(X_test_scaled.shape[0], X_test_scaled.shape[1], 1)
```

```python
model = KerasClassifier(build_fn=create_model, input_shape=(X_train_scaled.shape[1], 1), epochs=10, batch_size=32)
```

```python
train_scores = []
test_scores = []
histories = []

y_train_array = y_train.to_numpy()

kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation
for train_idx, test_idx in kfold.split(X_train_scaled, y_train_array):
    X_train_scaled_fold, X_test_scaled_fold = X_train_scaled[train_idx], X_train_scaled[test_idx]
    y_train_fold, y_test_fold = y_train_array[train_idx], y_train_array[test_idx]

    # Fit the model and capture the history
    history = model.fit(X_train_scaled_fold, y_train_fold, verbose=0)

    train_score = model.score(X_train_scaled_fold, y_train_fold)
    test_score = model.score(X_test_scaled_fold, y_test_fold)

    train_scores.append(train_score)
    test_scores.append(test_score)
    histories.append(history)  # Save the history

train_mean = np.mean(train_scores)
train_std = np.std(train_scores)
test_mean = np.mean(test_scores)
test_std = np.std(test_scores)
```

```python
# plot the accuracy
plt.plot(histories[0].history['accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train'], loc='upper left')
plt.show()
```

```python
def evaluate_model(y_true, y_pred, y_pred_proba):
    # Calculate metrics
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    auc_roc = roc_auc_score(y_true, y_pred_proba)

    # Print metrics
    print(f'Precision: {precision:.2f}')
    print(f'Recall: {recall:.2f}')
    print(f'F1 Score: {f1:.2f}')
    print(f'AUC-ROC: {auc_roc:.2f}')
    
    # Generate confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

    # Generate ROC curve
    fpr, tpr, _ = roc_curve(y_true, y_pred_proba)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f'AUC (Area = {auc_roc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()
```

```python
y_pred = model.predict(X_test_scaled)
y_pred_proba = model.predict_proba(X_test_scaled)[:, 1]

# Convert y_test to numpy array if it's not already
y_test_array = y_test.to_numpy()
```

```python
evaluate_model(y_test_array, y_pred, y_pred_proba)
```

Interpretation of the results with respect to the positive class (PTE) and negative class (ADE):

- Precision: 0.91
Precision represents the proportion of correctly identified PTE cases out of all predicted PTE cases. A precision of 0.91 means that out of all the cases predicted as PTE, 91% of them were actually PTE cases.

- Recall: 1.00
Recall, also known as sensitivity or true positive rate, measures the proportion of correctly identified PTE cases out of all actual PTE cases. A recall of 1.00 indicates that all positive PTE cases were correctly identified.

- F1 Score: 0.95
The F1 score is the harmonic mean of precision and recall, providing a balanced measure of model performance. An F1 score of 0.95 suggests a good balance between correctly identifying PTE cases and minimizing false positives.

- AUC-ROC: 0.63
The AUC-ROC evaluates the model's performance in distinguishing between the positive class (PTE) and negative class (ADE) across different probability thresholds. An AUC-ROC of 0.63 indicates a moderate discriminatory power in correctly classifying PTE cases compared to ADE cases.

In summary, the model shows high recall, indicating it correctly identifies all positive PTE cases. The precision is also high, suggesting a relatively low number of false positive predictions for PTE. The F1 score reflects a good balance between precision and recall. However, the AUC-ROC score of 0.63 indicates that the model's ability to discriminate between PTE and ADE cases is moderate.
