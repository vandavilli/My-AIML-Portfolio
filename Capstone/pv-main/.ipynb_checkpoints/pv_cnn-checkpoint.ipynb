```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, accuracy_score, precision_score, recall_score, f1_score
from gensim.models import KeyedVectors
from keras.preprocessing.text import Tokenizer
from keras.models import Sequential
from keras.layers import Dense, Embedding, Conv1D, MaxPooling1D, GlobalMaxPooling1D
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
```

```python
%run common_functions.ipynb
```

```python
train_data = pd.read_csv('data/train.csv')
test_data = pd.read_csv('data/test.csv')

dataframes = [train_data, test_data]
df = process_load_data(dataframes)
```

```python
model_path = './biowordvec/BioWordVec_PubMed_MIMICIII_d200.vec.bin'
bio_word_vec_model = load_wordvec_model(model_path)
```

```python
def word_2_vec(sentence):
    words = sentence.split()
    word_vecs = [bio_word_vec_model[word] for word in words if word in bio_word_vec_model]
    return np.mean(word_vecs, axis=0) if len(word_vecs) else np.zeros(bio_word_vec_model.vector_size)
```

```python
df['vec'] = df['context'].apply(word_2_vec)
data = np.stack(df['vec'].values)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, df['sentiment'], test_size=0.2, random_state=42)
```

```python
y_train = pd.Series(y_train)
y_train.reset_index(drop=True, inplace=True)
```

```python
def create_model():
    model = Sequential()
    model.add(Conv1D(128, 5, activation='relu', input_shape=(X_train.shape[1], 1)))
    model.add(MaxPooling1D(5))
    model.add(Conv1D(128, 5, activation='relu'))
    model.add(GlobalMaxPooling1D())
    model.add(Dense(128, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
model = KerasClassifier(build_fn=create_model, epochs=10, batch_size=32)
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
train_scores = []
test_scores = []

# Perform cross-validation
for train_idx, test_idx in kfold.split(X_train, y_train):
    X_train_fold, X_test_fold = X_train[train_idx], X_train[test_idx]
    y_train_fold, y_test_fold = y_train[train_idx], y_train[test_idx]

    model.fit(X_train_fold, y_train_fold, verbose=0)

    train_score = model.score(X_train_fold, y_train_fold)
    test_score = model.score(X_test_fold, y_test_fold)

    train_scores.append(train_score)
    test_scores.append(test_score)

train_mean = np.mean(train_scores)
train_std = np.std(train_scores)
test_mean = np.mean(test_scores)
test_std = np.std(test_scores)
```

```python
print("Cross-Validation Scores:")
print(test_scores)
print("Train Accuracy: {:.2f} +/- {:.2f}".format(train_mean, train_std))
print("Test Accuracy: {:.2f} +/- {:.2f}".format(test_mean, test_std))
```

```python
# Make predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Confusion matrices
train_cm = confusion_matrix(y_train, y_train_pred)
test_cm = confusion_matrix(y_test, y_test_pred)

print("Train Confusion Matrix:")
print(train_cm)
print("Test Confusion Matrix:")
print(test_cm)
```

### Conclusion (rewrite)
The Network was trained over 10 epochs with a batch size of 32. The accuracy of the model on the training set remained fairly constant at around 90.98% throughout all the epochs.

The cross-validation scores indicate the accuracy achieved by the model in each fold of the cross-validation process is fairly consistent

The scores are quite consistent across the folds, ranging from 0.909 to 0.916. This suggests that the model's performance is stable and not heavily influenced by the specific train-test splits. Overall, the model achieves an average accuracy of approximately 91.1% based on these cross-validation scores

The loss on the training set started at 0.3215 in the first epoch and gradually decreased to 0.2542 in the final epoch. This suggests that the model was learning and improving as it was trained over more epochs.

However, the validation loss decreased from 0.2970 in the first epoch to 0.2548 in the final epoch, which suggests that the model was not overfitting to the training data as it was able to maintain a low loss rate on unseen data.

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# Define class labels
class_labels = ['PTE', 'ADE']

# Confusion matrix for training data
sns.heatmap(train_cm, annot=True, ax=axs[0], cmap='Blues', fmt='g', 
            xticklabels=class_labels, yticklabels=class_labels)
axs[0].set_title('Train Confusion Matrix')
axs[0].set_xlabel('Predicted')
axs[0].set_ylabel('Actual')

# Confusion matrix for test data
sns.heatmap(test_cm, annot=True, ax=axs[1], cmap='Blues', fmt='g', 
            xticklabels=class_labels, yticklabels=class_labels)
axs[1].set_title('Validation Confusion Matrix')
axs[1].set_xlabel('Predicted')
axs[1].set_ylabel('Actual')

plt.show()
```

### Conclusion ( rewrite )

From the confusion matrix for both training and test data, it seems the model is not performing well on the minority class (class 0 - PTE).

Even though the overall accuracy of the model is about 91%, the precision, recall and f1-score for class 0 (PTE) is 0. This indicates that the model is unable to correctly identify any true positives for the class PTE, leading to low performance scores for this class. This was expected because the dataset had low number of PTE's and more of ADE's

On the other hand, the model performs well on the majority class (class 1 - ADE) with a precision, recall and f1-score close to 1. This is a clear case of class imbalance where the model is biased towards the majority class.

```python
from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score

# Predict probabilities for train and test set
y_train_prob = model.predict(X_train.reshape(*X_train.shape, 1))
y_test_prob = model.predict(X_test.reshape(*X_test.shape, 1))

# Calculate AUC-ROC
auc_roc_train = roc_auc_score(y_train, y_train_prob)
auc_roc_test = roc_auc_score(y_test, y_test_prob)

print(f'Train AUC-ROC: {auc_roc_train}')
print(f'Test AUC-ROC: {auc_roc_test}')

# Calculate Precision, Recall and F1-score for the positive class
precision_train = precision_score(y_train, y_train_pred)
recall_train = recall_score(y_train, y_train_pred)
f1_train = f1_score(y_train, y_train_pred)

precision_test = precision_score(y_test, y_test_pred)
recall_test = recall_score(y_test, y_test_pred)
f1_test = f1_score(y_test, y_test_pred)

print('\nTrain Precision: {:.2f}, Recall: {:.2f}, F1-score: {:.2f}'.format(precision_train, recall_train, f1_train))
print('Test Precision: {:.2f}, Recall: {:.2f}, F1-score: {:.2f}'.format(precision_test, recall_test, f1_test))
```

```python
### visualize Precision, Recall, and F1-score metrics

metrics = ['Precision', 'Recall', 'F1-score']
train_values = [precision_train, recall_train, f1_train]
test_values = [precision_test, recall_test, f1_test]

x = np.arange(len(metrics))

fig, ax = plt.subplots()
bar_width = 0.35
rects1 = ax.bar(x - bar_width/2, train_values, bar_width, label='Train')
rects2 = ax.bar(x + bar_width/2, test_values, bar_width, label='Test')

ax.set_ylabel('Scores')
ax.set_title('Scores by Metrics and Dataset')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

fig.tight_layout()
plt.show()
```

```python
# Calculate false positive rate and true positive rate for train and test set
fpr_train, tpr_train, _ = roc_curve(y_train, y_train_prob)
fpr_test, tpr_test, _ = roc_curve(y_test, y_test_prob)

plt.figure()
plt.plot(fpr_train, tpr_train, label='Train (AUC = {:.2f})'.format(auc_roc_train))
plt.plot(fpr_test, tpr_test, label='Test (AUC = {:.2f})'.format(auc_roc_test))
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

### Conclusion: (rewrite)

Area Under the Curve (AUC) score of 0.83 for both the training and test sets suggests that the model has a good performance in distinguishing between the positive and negative classes. It also indicates that the model is not overfitting since the performance on both the train and test sets is similar.
