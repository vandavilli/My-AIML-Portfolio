### kernel_length vs. groove_length

```python
klength = seeds_df['Kernel_Length']
glength = seeds_df['Groove_Length']
plt.plot(klength, glength, 'o', markersize=4, alpha=0.4)
plt.xlabel('Kernel Length')
plt.ylabel('Groove Length')
```

```python
from scipy.stats import linregress
res = linregress(seeds_df['Kernel_Length'], seeds_df['Groove_Length'])
res
```

```python
fx = np.array([seeds_df['Kernel_Length'].min(), seeds_df['Kernel_Length'].max()])
fy = res.intercept + res.slope * fx
plt.plot(fx, fy, '-')
plt.show()
```

### Logistic Regression

```python
seeds_df['Classification'] = seeds_df['Classification'].map({'Kama':1, 'Rosa':2, 'Canadian':3})
```

```python
seeds2_df = seeds_df[ (seeds_df.Classification==1) | (seeds_df.Classification==2) ]
```

### 2 classes (Kama & Rosa ) & 1 input - kernel_length

```python
X = seeds2_df[['Kernel_Length']]
y = seeds2_df['Classification']
```

```python
lr = LogisticRegression().fit(X,y)
```

```python
logreg_beta0 = lr.intercept_[0]
logreg_beta1 = lr.coef_[0,0]
logreg_thresh = -logreg_beta0/logreg_beta1

(logreg_beta0, logreg_beta1, logreg_thresh)
```

```python
X1 = X[y==1].to_numpy()
X2 = X[y==2].to_numpy()

mu1 = np.mean(X1)
mu2 = np.mean(X2)

var1 = np.var(X1)
var2 = np.var(X2)

var = np.mean((var1,var2))

approx_beta1 = (mu2-mu1)/var
approx_beta0 = (mu1**2 - mu2**2)/2/var
approx_thresh = -approx_beta0/approx_beta1

(approx_beta0, approx_beta1, approx_thresh)
```

```python
def sigmoid(beta0,beta1,x):
    return 1/(1+np.exp(-beta0 - beta1*x))

plt.clf()

x = np.linspace(3,7)
plt.figure(figsize=(8,6))
plt.scatter( X1, np.zeros(70), label='Kama' )
plt.scatter( X2, np.ones(70), label='Rosa' )
plt.plot(x, sigmoid(logreg_beta0,logreg_beta1,x), linewidth=4, color='green',   label='logistic regression')
plt.plot(x, sigmoid(approx_beta0,approx_beta1,x), linewidth=3, color='magenta', label='approximation')
plt.legend(fontsize=14)
plt.grid()

plt.show()
```

### 2 classes (Kama & Rosa ) & 2 inputs - kernel_length & groove_length

```python
X = seeds2_df[['Kernel_Length','Groove_Length']]
y = seeds2_df['Classification']
```

```python
lr = LogisticRegression().fit(X, y)
lr.intercept_, lr.coef_
```

```python
def plot_region(lr,d1lim,d2lim):
    d1_grid, d2_grid = np.meshgrid(np.arange(d1lim[0], d1lim[1], 0.02), np.arange(d2lim[0], d2lim[1], 0.02))
    d12_array = np.c_[d1_grid.ravel(), d2_grid.ravel()]
    y_array = lr.predict(d12_array)
    y_grid = y_array.reshape(d1_grid.shape)
    plt.contourf(d1_grid, d2_grid, y_grid, cmap='Pastel2')
```
