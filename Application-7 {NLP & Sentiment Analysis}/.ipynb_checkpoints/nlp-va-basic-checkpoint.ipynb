# Stemming using PorterStemmer() over Snowball & Lancaster

```python
# stemming function & removing stop words & punctuation
def stemmer(text):
    stemmer = PorterStemmer()
    swords = stopwords.words('english')
    punctuation = string.punctuation
    return ' '.join([stemmer.stem(w) for w in word_tokenize(text) 
                     if ((w not in swords) and (w not in punctuation))])
```

```python
stemmed_tokens = df['text'].apply(stemmer)
print(stemmed_tokens) ### Note how the readability gets compromized.
### Note: Leads to Over-stemming ( FP ) & Under-stemming ( FN)
```

# Lemmatizing

```python
# lemmatizing function & removing stop words
def lemmatizer(text):
    lemma = WordNetLemmatizer()
    swords = stopwords.words('english')
    punctuation = string.punctuation
    return ' '.join([lemma.lemmatize(w) for w in word_tokenize(text) if ((w not in swords) and (w not in punctuation))
                    ])
```

```python
lemmatized_tokens = df['text'].apply(lemmatizer)
print(lemmatized_tokens)
```

# Word Frequency Distribution

```python
### Preprocessing Function
import re
def preprocess_mesg(message):
    # define list to hold all the preprocessed words
    preprocessed_msg = [] 
    # first lets convert the messages into lower case
    message_lower = message.lower()
    message = re.sub(r'[^a-zA-Z0-9\s]', ' ', message_lower)
    # combine stopwords, and punctuations
    stopword_and_punctiation = set(stopwords.words('english') + list(string.punctuation))
    # tokenize each message with word_tokenize
    message = word_tokenize(message)
    # Initialize the lemmatizer
    lemmatizer = WordNetLemmatizer()
    # clean the message
    cleaned_message = [preprocessed_msg.append(lemmatizer.lemmatize(word)) for word in message if word not in stopword_and_punctiation]
    cleaned_message = " ".join(preprocessed_msg)
    
    return cleaned_message
                               
```

```python
df['text'] = df['text'].apply(preprocess_mesg)
```

```python
freq_distribution = FreqDist(df['text'])
freq_distribution.plot(20, cumulative=True)
```

```python
# grabbing the predictor & target
X = df['text'].values
y = df['humor'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Convert training samples into numerical data
tfidf = TfidfVectorizer()
tfidf_train = tfidf.fit_transform(X_train)
tfidf_test = tfidf.transform(X_test)
print(tfidf.get_feature_names_out()[:10])
```

# LogisticRegression

```python
logreg_pipe = Pipeline([('lr', LogisticRegression(random_state=42, max_iter = 1000))])

params_logreg = {
    'lr__penalty' : ['l2'], 
    'lr__C'       : np.logspace(-3,3,7),
    'lr__solver'  : ['lbfgs']
}
```

```python
def calculate_performance(pipe, params, X_train, y_train, X_test, y_test):
    grid = GridSearchCV(pipe, 
                        param_grid=params, 
                        scoring='accuracy').fit(X_train, y_train)
    
    return grid
```

```python
#logistic regression
lr_grid = calculate_performance(logreg_pipe, params_logreg, tfidf_train, y_train, tfidf_test, y_test)

lr_cv_results = lr_grid.cv_results_
```

# MultinomialNB

```python
nb_pipe = Pipeline([('nb', MultinomialNB())])

params_nb = {
    'nb__alpha' : np.arange(0, 1, .1)
}
```

```python
def train_and_predict(X,y):
    nb_grid = GridSearchCV(nb_pipe, 
                        param_grid=params_nb, 
                        scoring='accuracy').fit(X, y)
    return nb_grid
```

```python
nb_grid = train_and_predict(tfidf_train, y_train)
nb_cv_results = nb_grid.cv_results_
```

### Findings

```python
results_df = pd.DataFrame(
    {
        'Model': ['LogisticRegression','MultinomalNB'],
        'Best Params':[lr_grid.best_params_, nb_grid.best_params_],
        'Best Score':[lr_grid.score(tfidf_test, y_test), nb_grid.score(tfidf_test, y_test)]
    }
).set_index('Model')
```
